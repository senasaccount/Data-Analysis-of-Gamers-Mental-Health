**Data Cleaning**

```python
df_hours_over_100 = df[df["Hours"]>100]
df_hours_over_100["Hours"].count()
```

```python
index_hours_over_100 = df[df["Hours"] > 100].index
```

```python
df.drop(index_hours_over_100, inplace=True)
```

```python
df_streams_over_40 = df[df['streams']>40]
df_streams_over_40.streams.count()
```

```python
index_streams_over_100 = df[df["streams"] > 40].index
```

```python
df.drop(index_streams_over_100, inplace=True)
```

```python
threshold = len(df)*0.1
threshold
```

```python
df.dropna(thresh=threshold, axis=1, inplace=True)
```

```python
def impute_median(column):
  return column.fillna(column.median())
```

```python
df.Hours = df["Hours"].transform(impute_median)
df.streams = df["streams"].transform(impute_median)
df.SPIN_T = df["SPIN_T"].transform(impute_median) #mode alÄ±nacak
```

```python
df['GADE'].fillna(df['GADE'].mode().values[0], inplace=True)
df['League'].fillna(str(df['League'].mode().values[0]), inplace=True)
df['SPIN1'].fillna(df['SPIN1'].mode().values[0], inplace=True)
df['SPIN2'].fillna(df['SPIN2'].mode().values[0], inplace=True)
df['SPIN3'].fillna(df['SPIN3'].mode().values[0], inplace=True)
df['SPIN4'].fillna(df['SPIN4'].mode().values[0], inplace=True)
df['SPIN5'].fillna(df['SPIN5'].mode().values[0], inplace=True)
df['SPIN6'].fillna(df['SPIN6'].mode().values[0], inplace=True)
df['SPIN7'].fillna(df['SPIN7'].mode().values[0], inplace=True)
df['SPIN8'].fillna(df['SPIN8'].mode().values[0], inplace=True)
df['SPIN9'].fillna(df['SPIN9'].mode().values[0], inplace=True)
df['SPIN10'].fillna(df['SPIN10'].mode().values[0], inplace=True)
df['SPIN11'].fillna(df['SPIN11'].mode().values[0], inplace=True)
df['SPIN12'].fillna(df['SPIN12'].mode().values[0], inplace=True)
df['SPIN13'].fillna(df['SPIN13'].mode().values[0], inplace=True)
df['SPIN14'].fillna(df['SPIN14'].mode().values[0], inplace=True)
df['SPIN15'].fillna(df['SPIN15'].mode().values[0], inplace=True)
df['SPIN16'].fillna(df['SPIN16'].mode().values[0], inplace=True)
df['SPIN17'].fillna(df['SPIN17'].mode().values[0], inplace=True)
df['SPIN_T'].fillna(df['SPIN_T'].mode().values[0], inplace=True)
df['Narcissism'].fillna((df['Narcissism'].mode().values[0]), inplace=True)
df['Work'].fillna(str(df['Work'].mode().values[0]), inplace=True)
df['Reference'].fillna(str(df['Reference'].mode().values[0]), inplace=True)
df['accept'].fillna(str(df['accept'].mode().values[0]), inplace=True)
df['Residence_ISO3'].fillna(str(df['Residence_ISO3'].mode().values[0]), inplace=True)
df['Birthplace_ISO3'].fillna(str(df['Birthplace_ISO3'].mode().values[0]), inplace=True)
```

**Determining Outliers of Hours Column**



```python
df.Hours.quantile([0.25, 0.5, 0.75]) #analiz
```

```python
Q1_hours = df.Hours.quantile(0.25)
Q2_hours = df.Hours.quantile(0.50)
Q3_hours = df.Hours.quantile(0.75)
IQR_hours = Q3_hours - Q1_hours
IQR_hours
```

```python
minValueOfHours = Q1_hours - 1.5*IQR_hours
maxValueOfHours = Q3_hours + 1.5*IQR_hours
```

```python
minValueOfHours
```

```python
maxValueOfHours
```

```python
upper_outlier = df[df["Hours"]>=maxValueOfHours]
upper_outlier['Hours']
```

```python
df.boxplot(column="Hours")  # Here, we can clearly see the outliers of playing hours. As we determined it's starts with 52
```
